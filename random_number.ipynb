```python
import random
import pandas as pd
```

```python
df = pd.read_csv('Agent_list.csv')
```

```python
length = int(len(df)/2)
```

```python
df_1 = df.sample(frac=0.5,random_state=length)
```

```python
df_1
```

```python
df_2 = df.drop(df_1.index)
```

```python
df_2
```

```python
df_1 = df_1.reset_index(drop=True)
```

```python
df_1
```

```python
df_2 = df_2.reset_index(drop=True)
```

```python
df_2
```

```python
df_result = pd.merge(df_1,df_2,left_index=True, right_index=True)
```

```python
df_result
```

```python
df_result.to_csv('result.csv')
```
